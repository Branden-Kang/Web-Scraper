```python
import pandas as pd
import datetime as dt
import pandas_datareader.data as web
```

# Download Dataset


```python
start_time = dt.datetime(2000, 1, 1)
end_time = dt.datetime(2022, 2, 1)
df = web.DataReader('HISTPOP', 'oecd', start_time, end_time)
```

```python
print(df.shape)
df.info()
df.head()
```

```python
country = "Japan"
sex = "Total"  # Men and Women
df_japan_total = df[country][sex]
df_japan_total.head()
```
