```python
import pandas as pd
import numpy as np

from urllib.request import urlopen
from bs4 import BeautifulSoup

import re
```

```python
url = "http://www.hubertiming.com/results/2018CRFH"
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
title = soup.title
print(title)
text = soup.get_text()
```

```python
soup.find_all('a')
```

```python
all_links = soup.find_all("a")
for link in all_links:
    print(link.get("href"))
```

```python
rows = soup.find_all('tr')
print(rows[:10])
#We are printing the first 10 rows.
```

```python
for row in rows:
    row_td = row.find_all('td')
print(row_td)
type(row_td)
```

```python
str_cells = str(row_td)
cleantext = BeautifulSoup(str_cells, "lxml").get_text()
print(cleantext)
```

```python
#First, we need to import the regular expressions moduleimport re

list_rows = []
for row in rows:
    cells = row.find_all('td')
    str_cells = str(cells)
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean, '',str_cells))
    list_rows.append(clean2)
print(clean2)
type(clean2)
```

```python
df = pd.DataFrame(list_rows)
df.head(10)
```

```python
df1 = df[0].str.split(',', expand=True)
df1.head(10)
```
